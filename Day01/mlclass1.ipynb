What is AI?
AI (Artificial Intelligence) is the field of computer science focused on building machines or software that can mimic human intelligence. This includes the ability to:

Learn from data (like humans learn from experience)

Reason and make decisions

Understand language and communicate

Recognize images or patterns

Solve problems and adapt to changes

```python
Type	              Description	                    Status	                                    Example
Narrow AI	Task-specific, limited intelligence        	Exists today	                   Siri, Google Maps
General AI	Human-level intelligence, adaptable      	Not yet                   	      (Still in research)
Super AI	Exceeds human intelligence	                Hypothetical	                   Sci-fi AIs (Jarvis)
```

```python
What is Machine Learning (ML)?
Machine Learning is a subset of Artificial Intelligence (AI) that focuses on building systems that
learn from data — instead of being explicitly programmed.

Type	                 Description	                                 Example
Supervised Learning	     Learn from labeled data (input + output)	     Predict house prices, spam detection
Unsupervised Learning	 Learn from unlabeled data (find patterns)	     Customer segmentation, anomaly detection
Reinforcement Learning	 Learn by interacting with environment and
                         getting rewards/punishments	                 Game playing AI (e.g., AlphaGo), robotics
```

```python
What is Deep Learning (DL)?
Deep Learning is a subset of Machine Learning, which itself is a part of Artificial Intelligence. It uses artificial neural networks to 
mimic how the human brain works — especially for learning from large amounts of data.
```
